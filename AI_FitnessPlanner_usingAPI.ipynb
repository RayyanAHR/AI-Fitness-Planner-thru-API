```python
# PURPOSE:
# Import required libraries

from groq import Groq
```

```python
# AI Fitness Planner

# This project generates personalized workout and meal plans using:
# - Greedy Search Algo
# - AI-based recommendation system
# - Groq LLM API
# - Prompt Engineering
# - Workflow automation concepts
```

```python
# PURPOSE:
# Take user fitness information

goal = input("Enter fitness goal: ") # eg. Weight Loss, Muscle Gain, Endurance
fitness_level = input("Enter fitness level: ") #eg. Beginner
meal_preference = input("Enter meal preference: ")  # eg. High Protein, Balanced Diet, Low Carb

user_data = {
    "goal": goal,
    "fitness_level": fitness_level,
    "meal_preference": meal_preference
}

print(user_data)
```

```python
# PURPOSE:
# Store workout information

workouts = {

    "Weight Loss": [

        {
            "exercise": "Running",
            "duration": 45,
            "priority": 5,
            "difficulty": "Easy"
        },

        {
            "exercise": "Cycling",
            "duration": 30,
            "priority": 4,
            "difficulty": "Easy"
        },

        {
            "exercise": "Jump Rope",
            "duration": 20,
            "priority": 3,
            "difficulty": "Medium"
        }
    ],

    "Muscle Gain": [

        {
            "exercise": "Weight Training",
            "duration": 60,
            "priority": 5,
            "difficulty": "Medium"
        },

        {
            "exercise": "Push Ups",
            "duration": 30,
            "priority": 4,
            "difficulty": "Easy"
        }
    ],

    "Endurance": [

        {
            "exercise": "Jogging",
            "duration": 40,
            "priority": 5,
            "difficulty": "Easy"
        },

        {
            "exercise": "Swimming",
            "duration": 45,
            "priority": 4,
            "difficulty": "Medium"
        }
    ]
}
```

```python
# PURPOSE:
# Store meal plans

meals = {

    "High Protein": [
        "Eggs",
        "Chicken Breast",
        "Protein Shake",
        "Rice and Vegetables"
    ],

    "Balanced Diet": [
        "Rice",
        "Salad",
        "Fruits",
        "Chicken"
    ],

    "Low Carb": [
        "Fish",
        "Vegetables",
        "Nuts",
        "Boiled Eggs"
    ]
}
```

```python
# PURPOSE:
# Generate workout plan using GREEDY SEARCH algorithm

def generate_workout_plan(goal, fitness_level):

    selected_workouts = workouts[goal]

    filtered_workouts = []

    for workout in selected_workouts:

        if fitness_level == "Beginner":

            if workout["difficulty"] != "Hard": # keep only easy and med as diff
                filtered_workouts.append(workout)

        else:
            filtered_workouts.append(workout)

    sorted_workouts = sorted( # concept of greedy so it only prefers higher priority ones first, for best exercises
        filtered_workouts,
        key=lambda x: x["priority"],
        reverse=True
    )

    return sorted_workouts
```

```python
# PURPOSE:
# Generate initial workout plan

initial_workout_plan = generate_workout_plan(
    user_data["goal"],
    user_data["fitness_level"]
)

print("INITIAL WORKOUT PLAN\n")

for workout in initial_workout_plan:

    print(
        f"{workout['exercise']} "
        f"- {workout['duration']} mins"
    )
```

```python
# PURPOSE:
# Display meal plan

print("MEAL PLAN\n")

for meal in meals[user_data["meal_preference"]]:

    print(meal)
```

```python
# PURPOSE:
# Configure Groq API

import os
from groq import Groq

client = Groq(api_key=os.getenv("GROQ_API_KEY"))
```

```python
# PURPOSE:
# Convert plan into text

plan_text = "Workout Plan:\n"

for workout in initial_workout_plan:

    plan_text += (
        f"- {workout['exercise']} "
        f"({workout['duration']} mins)\n"
    )

plan_text += "\nMeal Plan:\n"

for meal in meals[user_data["meal_preference"]]:

    plan_text += f"- {meal}\n"

print(plan_text)
```

```python
# PURPOSE:
# Create AI prompt (prompt engineering)

# in this AI will receive workout and meal plans, and will improve them
prompt = f"""
You are an intelligent fitness assistant.

Here is a fitness plan:

{plan_text}

Improve this plan by:
- Optimizing workout structure
- Suggesting healthier meals
- Making recommendations beginner-friendly
- Giving short fitness advice

Return the improved plan clearly.
"""
```

```python
# PURPOSE:
# Generate AI-improved fitness plan

response = client.chat.completions.create(  # for sending prompt to llama

    model="llama-3.1-8b-instant",

    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)

improved_plan = response.choices[0].message.content     # extracts the AI response text

print(improved_plan)
```

```python
# PURPOSE:
# Ask AI to explain recommendations

explanation_prompt = f"""
Explain why this fitness plan is effective:

{improved_plan}
"""

response2 = client.chat.completions.create(

    model="llama-3.1-8b-instant",

    messages=[
        {
            "role": "user",
            "content": explanation_prompt
        }
    ]
)

print(response2.choices[0].message.content)
```

```python
# PURPOSE:
# Handle invalid inputs

if goal not in workouts:
    print("Invalid fitness goal selected")
```

```python
# PURPOSE:
# Simulate workflow automation

print("Collecting user data...")
print("Applying search algorithm...")
print("Generating AI recommendations...")
print("Preparing final fitness plan...")
```
